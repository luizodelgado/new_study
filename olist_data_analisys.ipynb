```python
status_qty = db.groupby("order_status")["order_id"].nunique().reset_index()

status_qty.columns = ["order_status", "order_qty"]

status_qty
```

```python
db_transit_time = db[db["order_status"]== "delivered"].copy()

db_transit_time["order_delivered_customer_date"] = pd.to_datetime(db_transit_time["order_delivered_customer_date"])

db_transit_time["order_purchase_timestamp"] = pd.to_datetime(db_transit_time["order_purchase_timestamp"])

db_transit_time["transit_time"] = (db_transit_time["order_delivered_customer_date"]-db_transit_time["order_purchase_timestamp"]).dt.days

db_transit_time["transit_time"] = db_transit_time["transit_time"].astype("Int64")

avg_transit_time = db_transit_time["transit_time"].mean()

print("Average transit time: ", avg_transit_time)
```

```python
db_month_year = db_transit_time.copy()

db_month_year["month"] = db_transit_time["order_purchase_timestamp"].dt.month.astype(str)

db_month_year["year"] = db_transit_time["order_purchase_timestamp"].dt.year.astype(str)

db_month_year["month_year_str"] = db_month_year["month"] + "_" + db_month_year["year"]

db_month_year["month_year_to_period"] = db_transit_time["order_purchase_timestamp"].dt.to_period("M")

db_month_year["month_year_date_trunc"] = db_month_year["month_year_to_period"].dt.to_timestamp()

db_month_year
```

```python
db_month_year_qty = db_month_year.groupby("month_year_to_period")["order_id"].nunique().reset_index()

db_month_year_qty.sort_values("order_id", ascending=False, inplace=True)

db_month_year_qty
```
